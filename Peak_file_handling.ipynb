```python
import pandas as pd
import os
import glob

# Same paths
DATASET_PATH = '/content/New Finance set (1).xlsx'
VIDEO_FOLDER = '/content/drive/MyDrive/IITP_Internship/merged_folder'

# Load dataset uni_ids
df = pd.read_excel(DATASET_PATH)
uni_ids = set(df['uni_id'].astype(str).str.strip())

# Get all MP4 filenames (without .mp4)
videos = glob.glob(os.path.join(VIDEO_FOLDER, "*.mp4"))
video_bases = {os.path.splitext(os.path.basename(v))[0] for v in videos}

# FIXED: Extra videos (set - set)
extra_videos_set = video_bases - uni_ids
extra_videos = sorted(list(extra_videos_set))

print(f"📊 Dataset uni_ids: {len(uni_ids)}")
print(f"🎥 Total MP4s: {len(video_bases)}")
print(f"✅ Dataset matches: {len(uni_ids)}")  # All dataset IDs present
print(f"❌ Extra videos: {len(extra_videos)} ({len(extra_videos)/len(video_bases)*100:.1f}%)")

print("\n📋 First 20 extra videos:")
for vid in extra_videos[:20]:
    print(f"- {vid}")
if len(extra_videos) > 20:
    print(f"... and {len(extra_videos)-20} more")

# Save full list with full paths
extra_with_paths = [v + '.mp4' for v in extra_videos]
pd.DataFrame({'extra_uni_id': extra_videos, 'full_filename': extra_with_paths}).to_csv('/content/extra_videos.csv', index=False)
print(f"\n💾 Saved: /content/extra_videos.csv ({len(extra_videos)} rows)")
```

## Create Extra Videos Folder

### Subtask:
Define a new target folder path for the extra videos and create this folder in your Google Drive if it does not already exist.


**Reasoning**:
To create the designated folder for extra videos, I will define the folder path and use the `os.makedirs` function with `exist_ok=True` to ensure it is created without error if it doesn't already exist. I will also remount the drive to ensure access to Google Drive.



```python
from google.colab import drive
import os

drive.mount('/content/drive')

# 1. Define the full path for the new folder named extra_videos_folder
extra_videos_folder_path = '/content/drive/MyDrive/IITP_Internship/extra_videos_folder'

# 2. Use os.makedirs() with exist_ok=True to create the folder
os.makedirs(extra_videos_folder_path, exist_ok=True)

print(f"Created or ensured existence of folder: {extra_videos_folder_path}")
```

**Reasoning**:
Now that the `extra_videos_folder` has been created, the next step is to move the identified 'extra' video files from the `merged_folder` into this new folder. I will iterate through the `extra_videos` list and use `shutil.move` to transfer each file.



```python
import shutil
import os

# Assuming extra_videos, VIDEO_FOLDER, and extra_videos_folder_path are already defined
# from previous steps.

# Ensure the target folder exists (already handled in previous step, but good to be safe)
os.makedirs(extra_videos_folder_path, exist_ok=True)

moved_count = 0
failed_count = 0

print(f"Moving {len(extra_videos)} extra videos to: {extra_videos_folder_path}")

for video_basename in extra_videos:
    # Construct the full path of the source file (in merged_folder)
    source_file_name = video_basename + '.mp4'
    source_path = os.path.join(VIDEO_FOLDER, source_file_name)

    # Construct the full path of the destination file (in extra_videos_folder)
    destination_path = os.path.join(extra_videos_folder_path, source_file_name)

    if os.path.exists(source_path):
        try:
            shutil.move(source_path, destination_path)
            moved_count += 1
            print(f"Moved: {source_file_name}")
        except Exception as e:
            failed_count += 1
            print(f"Failed to move {source_file_name}: {e}")
    else:
        failed_count += 1
        print(f"Source file not found: {source_file_name}")

print(f"\nOperation complete. Successfully moved: {moved_count}, Failed/Not found: {failed_count}")
```

## Verify Moved Videos

### Subtask:
List the contents of the new 'extra videos' folder to confirm that the videos have been successfully moved.


**Reasoning**:
To verify that the videos have been successfully moved, I will list the contents of the `extra_videos_folder_path`, print the total number of files, and then list each filename.



```python
import os

# extra_videos_folder_path is already defined from previous steps
# extra_videos_folder_path = '/content/drive/MyDrive/IITP_Internship/extra_videos_folder'

print(f"Verifying contents of: {extra_videos_folder_path}")

if os.path.exists(extra_videos_folder_path):
    # Get all files and directories in the folder
    contents = os.listdir(extra_videos_folder_path)

    # Filter to get only files
    files = [f for f in contents if os.path.isfile(os.path.join(extra_videos_folder_path, f))]

    print(f"Total files found: {len(files)}")
    if files:
        print("Files in folder:")
        for file_name in sorted(files):
            print(f"- {file_name}")
    else:
        print("Folder is empty.")
else:
    print("Error: Folder does not exist.")
```
