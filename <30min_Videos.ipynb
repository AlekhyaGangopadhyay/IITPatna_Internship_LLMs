Checking for <30 minute videos.

```python
!pip install moviepy
import os
from moviepy.editor import VideoFileClip  # ← FIXED: moviepy.editor
from google.colab import drive

drive.mount('/content/drive')

VIDEO_FOLDER = '/content/drive/MyDrive/IITP_Internship/merged_folder'
THIRTY_MINUTES = 30 * 60
VIDEO_EXTENSIONS = (".mp4", ".mkv", ".avi", ".mov", ".flv", ".wmv")

total_duration = 0
videos_over_30 = []

print("Scanning videos...\n")

for filename in os.listdir(VIDEO_FOLDER):
    if filename.lower().endswith(VIDEO_EXTENSIONS):
        file_path = os.path.join(VIDEO_FOLDER, filename)

        try:
            with VideoFileClip(file_path) as video:
                duration = video.duration
                total_duration += duration

                print(f"{filename} → {duration/60:.2f} minutes")

                if duration > THIRTY_MINUTES:
                    videos_over_30.append(filename)

        except Exception as e:
            print(f"Error processing {filename}: {e}")

print("\nTotal Duration:")
print(f"{total_duration/3600:.2f} hours")

print("\nVideos longer than 30 minutes:")
for v in videos_over_30:
    print(v)
```
